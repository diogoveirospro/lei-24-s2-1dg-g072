# US11 - Inquérito aos utilizadores do parque

## 1. Introdução

A $\textbf{User Story (US11)}$ tem como objetivo principal possibilitar a coleta de dados por meio do portal do usuário para analisar o uso do parque por diferentes faixas etárias. Uma pesquisa com três perguntas foi incorporada ao portal para entender o comportamento e as opiniões dos frequentadores em relação ao parque.

As três perguntas da pesquisa são:

* **Faixa Etária:** Os usuários são categorizados em três grupos: crianças (até 15 anos), adultos (entre 16 e 65 anos) e seniores (mais de 65 anos).
* **Recomendação do Parque:** Os participantes indicam se recomendariam o parque a outras pessoas, respondendo "Sim" (Y) ou "Não" (N).
* **Frequência de Visitas:** Os usuários informam quantas vezes frequentam o parque por mês, dando um valor numérico.

Com base nas respostas registradas no arquivo "Inquiry.csv", esta funcionalidade visa:

* Identificar o tipo de cada variável, categorizando-as como categóricas ou numéricas.
* Calcular a proporção de usuários de cada faixa etária que recomendam o parque a outras pessoas.
* Gerar boxplots individuais para cada faixa etária, exibindo a distribuição da frequência mensal de visitas e permitindo insights sobre o comportamento dos diferentes grupos etários em relação ao uso do parque.

O propósito final desta user story é facilitar uma compreensão mais abrangente do perfil dos usuários do parque, possibilitando a adaptação dos serviços oferecidos com base nas necessidades e preferências específicas de cada faixa etária.

## 2. Código e Resultados

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulando a carga de dados diretamente para exemplificação
from io import StringIO

data = """
Escalao,Y/N,Visits
2,Y,7
3,N,4
2,N,1
3,Y,4
3,Y,10
1,N,4
2,Y,6
3,N,8
3,N,7
"""
dados = pd.read_csv(StringIO(data))

# Renomear colunas para garantir consistência
dados.columns = ['Faixa_etaria', 'Recomendaria', 'Visitas_por_mes']

# Remover linhas com quaisquer valores nulos
dados.dropna(inplace=True)

# Checar se há algum dado nulo
print("Verificação de dados nulos:", dados.isnull().sum())

# Mapeamento das faixas etárias
faixas = {1: 'Criança (até 15 anos)', 2: 'Adulto (16 a 65 anos)', 3: 'Sénior (mais de 65 anos)'}
dados['Faixa_etaria'] = dados['Faixa_etaria'].map(faixas)

# Verificação dos dados após o mapeamento
print("Verificação após mapeamento:")
print(dados)

# Contagem de valores em cada grupo
print("Contagem por grupo:")
print(dados.groupby('Faixa_etaria').size())

# Checar se alguma categoria está vazia
if dados['Faixa_etaria'].isnull().any():
    print("Há faixas etárias não mapeadas ou nulas.")
else:
    # Gerar boxplot para cada faixa etária com relação ao número de visitas por mês
    plt.figure(figsize=(10, 6))
    dados.boxplot(column='Visitas_por_mes', by='Faixa_etaria')
    plt.title('Frequência de Visitas ao Parque por Faixa Etária')
    plt.suptitle('')
    plt.xlabel('Faixa Etária')
    plt.ylabel('Número de Visitas por Mês')
    plt.show()
```

A análise dos dados obtidos da User Story (US11) oferece insights valiosos sobre o comportamento e as opiniões dos frequentadores do parque, segmentados por faixas etárias distintas. Vamos explorar os principais pontos observados:

### Proporção de Recomendação do Parque por Faixa Etária
Ao calcular a proporção de usuários que recomendam o parque a outras pessoas, podemos observar como diferentes grupos etários percebem a experiência no parque.

Esses números destacam a satisfação relativa de cada grupo com o parque, sendo úteis para avaliar áreas que podem ser melhoradas ou mantidas conforme as preferências dos diferentes segmentos de público.

Dentre os destaques dessas análises, podemos identificar padrões de comportamento, como frequências médias de visitas, variações na distribuição e possíveis diferenças nas preferências de uso entre os grupos etários.
