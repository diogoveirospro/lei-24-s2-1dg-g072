# US9 - Custos do Consumo de Água

## 1. Introdução

$\textbf{Nesta User Story (US09)}$ foi-nos pedido a realização de uma análise estatística relativa aos custos de consumo de água em todos os parques. Recebemos um arquivo denominado $\textit{water_consumption.csv}$ que fornece os dados necessários para a realização do estudo. 
Este arquivo regista o consumo diário de água (em $m^3$) desde o dia de abertura de cada parque. O valor pago pela água é de $0.7\,€/m^3$ até um consumo de $1000\,m^3/mês$, com uma taxa de $15\%$ para consumos superiores.




## 2. Código e Resultados

```python
# importação das bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as st
import numpy as np
from scipy import stats
```

### Consumo Mensal de Água

```python
# leitura do ficheiro com o consumo de água
data = pd.read_csv('water_consumption.csv', sep=';', skiprows = 1, names=['Park', 'Year', 'Month', 'Day', 'Consumption'])

data['Month'] = data['Month'].astype(int)
data['Year'] = data['Year'].astype(int)
data['Consumption'] = data['Consumption'].astype(str).str.replace(',', '.').astype(float)
```

### Barplot

```python
ano = int(input('Digite o ano para o qual pretende o estudo: '))
mesInicio = int(input('Digite o mês no qual pretende começar o estudo: '))
mesFim = int(input('Digite o mês no qual pretende terminar o estudo: '))
identificacao_parque = (input('Digite a identificação do parque para o qual pretende o estudo: '))

dados_filtrados = data[(data['Year'] == ano) & 
                       (data['Month'] >= mesInicio) & 
                       (data['Month'] <= mesFim) & 
                       (data['Park'] == identificacao_parque)]

consumo_mensal = dados_filtrados.groupby('Month')['Consumption'].sum().reset_index()

sns.barplot(x='Month', y='Consumption', data=consumo_mensal, color='blue')
plt.title(f'Consumo Mensal de Água no Parque {identificacao_parque}')
plt.xlabel('Mês')
plt.ylabel('Consumo')
plt.show()
```

### Média dos Custos Mensais do Consumo de Água

```python
numero_parques = int(input('Digite o número de parques para os quais pretende a média mensal do consumo de água: '))

taxa_base = 0.7  # Taxa base por metro cúbico
limite_superior = 1000 # Limite de consumo para aplicar a taxa base

consumo_mensal = list()
custo_por_mes = list()
for i in range(numero_parques):
    identificacao_parque = input('Digite a identificação do parque para o qual pretende o estudo: ')
    
    for j in range(1, 13):
        dados_filtrados = data[(data['Park'] == identificacao_parque) & (data['Month'] == j)]
        consumo_total = dados_filtrados['Consumption'].sum()
        consumo_mensal.append(consumo_total)
    for k, valor in enumerate(consumo_mensal):
        if valor <= 1000:
            custo_por_mes.append(valor * taxa_base)
        else:
            custo_por_mes.append(valor * (taxa_base + (0.15 * taxa_base)))
            
    media_custo_mensal = st.mean(custo_por_mes)

    print(f'\nA Média de Custo Mensal de Consumo de Água no Parque {identificacao_parque} é de {media_custo_mensal:.2f}€\n')
```

### Comparar o Menor e Maior Parque em Termos de Consumo

```python
# Filtrar os dados para excluir os valores zero
data_filtrado = data[data['Consumption'] > 0]

# Encontrar o menor consumo geral, excluindo os valores zero
consumo_min = data_filtrado['Consumption'].min()
consumo_max = data['Consumption'].max()

# Determinar os parques correspondentes aos valores mínimo e máximo
parques_consumo_min = data.loc[data['Consumption'] == consumo_min, 'Park'].tolist()
parques_consumo_max = data.loc[data['Consumption'] == consumo_max, 'Park'].tolist()

# Calcular os indicadores estatísticos
def calcular_indicadores(dados):
    media = np.mean(dados)
    mediana = np.median(dados)
    desvio_padrao = np.std(dados)
    coeficiente_assimetria = stats.skew(dados)
    return [media, mediana, desvio_padrao, coeficiente_assimetria]

# Função para construir tabelas de frequência
def construir_tabela_frequencia(dados, classes=5):
    frequencia_absoluta, intervalos = np.histogram(dados, bins=classes)
    frequencia_relativa = frequencia_absoluta / len(dados)
    tabela_frequencia = pd.DataFrame({
        'Intervalo': [f'{intervalos[i]:.2f} - {intervalos[i+1]:.2f}' for i in range(len(intervalos) - 1)],
        'Frequência Absoluta': frequencia_absoluta,
        'Frequência Relativa': frequencia_relativa
    })
    return tabela_frequencia

# Verificar outliers
def verificar_outliers(dados):
    quartis = np.percentile(dados, [25, 50, 75])
    iqr = quartis[2] - quartis[0]
    limite_inferior = quartis[0] - 1.5 * iqr
    limite_superior = quartis[2] + 1.5 * iqr
    outliers = dados[(dados < limite_inferior) | (dados > limite_superior)]
    return outliers

# Parques com o menor consumo
for parque in set(parques_consumo_min):  # Usando set() para evitar duplicatas
    # Recolha dos dados do parque
    dados_parque_menor = data.loc[data['Park'] == parque, 'Consumption'].dropna()
    
    # Calcular indicadores estatísticos
    indicadores_parque_menor = calcular_indicadores(dados_parque_menor)

    # Construir tabela de frequência
    tabela_frequencia_parque_menor = construir_tabela_frequencia(dados_parque_menor)

    # Verificar os outliers
    outliers_parque_menor = verificar_outliers(dados_parque_menor)

    # Histogramas com 10 e 100 classes
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 2, 1)
    sns.histplot(dados_parque_menor, bins=10, kde=True, color='blue')
    plt.title(f'Histograma - {parque} (10 classes)')
    plt.subplot(2, 2, 2)
    sns.histplot(dados_parque_menor, bins=100, kde=True, color='blue')
    plt.title(f'Histograma - {parque} (100 classes)')
    plt.show()

    # Output dos resultados
    print(f'Parque com menor consumo de água ({parque}):')
    print(f'Média: {indicadores_parque_menor[0]:.2f}')
    print(f'Mediana: {indicadores_parque_menor[1]:.2f}')
    print(f'Desvio Padrão: {indicadores_parque_menor[2]:.2f}')
    print(f'Coeficiente de Assimetria: {indicadores_parque_menor[3]:.2f}\n')
    print('Tabela de Frequência:')
    print(tabela_frequencia_parque_menor)
    print(f'\nOutliers: {outliers_parque_menor.tolist()}\n')

# Parques com o maior consumo
for parque in set(parques_consumo_max):  # Usando set() para evitar duplicatas
    
    if parque in parques_consumo_min:
        print(f'O parque {parque} também tinha o consumo mínimo, por isso foi analisado estatisticamente anteriomente!')
    else:

        # Recolha dos dados do parque
        dados_parque_maior = data.loc[data['Park'] == parque, 'Consumption'].dropna()
    
        # Calcular indicadores estatísticos
        indicadores_parque_maior = calcular_indicadores(dados_parque_maior)

        # Construir tabela de frequência
        tabela_frequencia_parque_maior = construir_tabela_frequencia(dados_parque_maior)

        # Verificar os outliers
        outliers_parque_maior = verificar_outliers(dados_parque_maior)

        # Histogramas com 10 e 100 classes
        plt.figure(figsize=(12, 6))
        plt.subplot(2, 2, 1)
        sns.histplot(dados_parque_maior, bins=10, kde=True, color='blue')
        plt.title(f'Histograma - {parque} (10 classes)')
        plt.subplot(2, 2, 2)
        sns.histplot(dados_parque_maior, bins=100, kde=True, color='blue')
        plt.title(f'Histograma - {parque} (100 classes)')
        plt.show()

        # Output dos resultados
        print(f'Parque com maior consumo de água ({parque}):')
        print(f'Média: {indicadores_parque_maior[0]:.2f}')
        print(f'Mediana: {indicadores_parque_maior[1]:.2f}')
        print(f'Desvio Padrão: {indicadores_parque_maior[2]:.2f}')
        print(f'Coeficiente de Assimetria: {indicadores_parque_maior[3]:.2f}\n')
        print('Tabela de Frequência:')
        print(tabela_frequencia_parque_maior)
        print(f'\nOutliers: {outliers_parque_maior.tolist()}\n')
```

## 3. Análise e Interpretação dos Resultados

Na representação gráfica do $\textbf{custo mensal da água}$, podemos observar que o consumo mensal é mais elevado durante os meses de verão, quando há maior seca. Por outro lado, o consumo é relativamente menor durante os períodos mais húmidos, como na primavera, outono e inverno.

Após reunir os valores necessários para calcular o custo mensal, determinamos o preço a pagar multiplicando o valor do consumo pela taxa, resultando no montante a pagar. É importante notar que, dependendo se o consumo ultrapassar ou não os $1000 m^{3}$, o valor a pagar pode aumentar em até $15\%$ para consumos superiores a esse limiar.

A fórmula para calcular a mediana é:
Se $n$ for ímpar:
$\widetilde{x} = x_{\left(\frac{n + 1}{2}\right)}$, 
Se $n$ for par:
$\widetilde{x} = \frac{x_{\left(\frac{n}{2}\right)} + x_{\left(\frac{n}{2} + 1\right)}}{2}$,
para dados não classificados. Além disso criamos uma tabela de frequencias que serviu como base de visualização para a localização dos valores em frequencia absoluta e relativa, que ajudou a indentificar onde acontece o maior consumo diário, durante cada dia.

Por fim criamos histogramas para de 100 e 10 classes para nos ajudar a perceber melhor as informações existentes nas tabelas de frequencia, de uma forma mais geral para o de 10 classes e máis especidica para o de 100 classes. (A tabela estava dividida em quatro classes).

Sendo que assim entendemos melhor o estudo realizado para o consumo do maior e menor parque em termos do consumo diário e acabamos por observar que a média, mediana e desvio padrão são maiores no parque com maior consumo sendo o de assimetria ligeiramente maior, além de ambos apresentarem histogramas semelhantes com a unica diferença sendo o valor do de maior consumo ser normalmente maior do que o de menor.


# US10 - Uso de Equipamentos no Parque

## Introdução

$\textbf{Nesta User Story (US10)}$ foi-nos pedido a realização de uma análise estatística relativa aos equipamentos usados no parque durante o dia. Recebemos um arquivo denominado $\textit{Inquiry.csv}$ que fornece os dados necessários para a realização do estudo. 
Este arquivo regista as informações referentes as peças usadas durante o dia. Os equipamentos usados serão cerca de 1000 e o objetivo é criar uma piechart com as informações relativas a esse ficheiro em percentagem.

## Código e Resultados

```python
# importação das bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# leitura do ficheiro
data = pd.read_csv('EquipmentUsed.csv', sep=';', skiprows = 1, names=['Equipment'])
```

```python
equipamentos = list()
usoEquipamentos = []

equipamentos = data['Equipment'].unique().tolist()

for indice, equipamento in enumerate(equipamentos):
    contador = (data['Equipment'] == equipamento).sum()
    usoEquipamentos.append(contador)

plt.pie(usoEquipamentos, labels = equipamentos, autopct = '%1.1f%%', shadow = True, startangle = 90, 
        colors = sns.color_palette('muted'))

plt.axis('equal')
```

## Análise e Interpretação de Resultados

Através da representação grafica gerada do piechart conseguimos concluir que os equipamentos/locais mais usados são os $\textbf{Walking}$ $\textbf{trails}$ (Trilhasde caminhada) com uma percentagem de uso de 21.9%, de seguida são os $\textbf{picnic}$ $\textbf{area}$(areas de picnic) com 18.1%, os $\textbf{rest}$ $\textbf{benches}$(bancos de descanso) com 17.6%, o $\textbf{children's}$ $\textbf{playground}$(parque para crianças) com 13.9%, os $\textbf{toilletes}$(casas de banho) com 11.3%, as $\textbf{drinking}$ $\textbf{foutains}$(bebedouros) com 10.1% e por ultimo as $\textbf{exercise}$ $\textbf{machines}$(maquinas de exercicio) com 7.1 %.

# US11 - Inquérito aos utilizadores do parque

## Introdução

A $\textbf{User Story (US11)}$ tem como objetivo principal possibilitar a coleta de dados por meio do portal do usuário para analisar o uso do parque por diferentes faixas etárias. Uma pesquisa com três perguntas foi incorporada ao portal para entender o comportamento e as opiniões dos frequentadores em relação ao parque.

As três perguntas da pesquisa são:

* **Faixa Etária:** Os usuários são categorizados em três grupos: crianças (até 15 anos), adultos (entre 16 e 65 anos) e seniores (mais de 65 anos).
* **Recomendação do Parque:** Os participantes indicam se recomendariam o parque a outras pessoas, respondendo "Sim" (Y) ou "Não" (N).
* **Frequência de Visitas:** Os usuários informam quantas vezes frequentam o parque por mês, dando um valor numérico.

Com base nas respostas registradas no arquivo "Inquiry.csv", esta funcionalidade visa:

* Identificar o tipo de cada variável, categorizando-as como categóricas ou numéricas.
* Calcular a proporção de usuários de cada faixa etária que recomendam o parque a outras pessoas.
* Gerar boxplots individuais para cada faixa etária, exibindo a distribuição da frequência mensal de visitas e permitindo insights sobre o comportamento dos diferentes grupos etários em relação ao uso do parque.

O propósito final desta user story é facilitar uma compreensão mais abrangente do perfil dos usuários do parque, possibilitando a adaptação dos serviços oferecidos com base nas necessidades e preferências específicas de cada faixa etária.

## 2. Código e Resultados

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulando a carga de dados diretamente para exemplificação
from io import StringIO

data = """
Escalao,Y/N,Visits
2,Y,7
3,N,4
2,N,1
3,Y,4
3,Y,10
1,N,4
2,Y,6
3,N,8
3,N,7
"""
dados = pd.read_csv(StringIO(data))

# Renomear colunas para garantir consistência
dados.columns = ['Faixa_etaria', 'Recomendaria', 'Visitas_por_mes']

# Remover linhas com quaisquer valores nulos
dados.dropna(inplace=True)

# Checar se há algum dado nulo
print("Verificação de dados nulos:", dados.isnull().sum())

# Mapeamento das faixas etárias
faixas = {1: 'Criança (até 15 anos)', 2: 'Adulto (16 a 65 anos)', 3: 'Sénior (mais de 65 anos)'}
dados['Faixa_etaria'] = dados['Faixa_etaria'].map(faixas)

# Verificação dos dados após o mapeamento
print("Verificação após mapeamento:")
print(dados)

# Contagem de valores em cada grupo
print("Contagem por grupo:")
print(dados.groupby('Faixa_etaria').size())

# Checar se alguma categoria está vazia
if dados['Faixa_etaria'].isnull().any():
    print("Há faixas etárias não mapeadas ou nulas.")
else:
    # Gerar boxplot para cada faixa etária com relação ao número de visitas por mês
    plt.figure(figsize=(10, 6))
    dados.boxplot(column='Visitas_por_mes', by='Faixa_etaria')
    plt.title('Frequência de Visitas ao Parque por Faixa Etária')
    plt.suptitle('')
    plt.xlabel('Faixa Etária')
    plt.ylabel('Número de Visitas por Mês')
    plt.show()
```

## 3. Análise e Interpretação de Resultados

A análise dos dados obtidos da User Story (US11) oferece insights valiosos sobre o comportamento e as opiniões dos frequentadores do parque, segmentados por faixas etárias distintas. Vamos explorar os principais pontos observados:

### Proporção de Recomendação do Parque por Faixa Etária
Ao calcular a proporção de usuários que recomendam o parque a outras pessoas, podemos observar como diferentes grupos etários percebem a experiência no parque.

Esses números destacam a satisfação relativa de cada grupo com o parque, sendo úteis para avaliar áreas que podem ser melhoradas ou mantidas conforme as preferências dos diferentes segmentos de público.

### Boxplots da Frequência de Visitas por Faixa Etária
Os boxplots gerados para cada faixa etária em relação à frequência de visitas mensais fornecem uma visão clara da distribuição e dos hábitos de uso do parque.

Dentre os destaques dessas análises, podemos identificar padrões de comportamento, como frequências médias de visitas, variações na distribuição e possíveis diferenças nas preferências de uso entre os grupos etários.
