```python
# importação das bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# leitura do ficheiro
data = pd.read_csv('EquipmentUsed.csv', sep=';', skiprows = 1, names=['Equipment'])
```

```python
equipamentos = list()
usoEquipamentos = []

equipamentos = data['Equipment'].unique().tolist()

for indice, equipamento in enumerate(equipamentos):
    contador = (data['Equipment'] == equipamento).sum()
    usoEquipamentos.append(contador)

plt.pie(usoEquipamentos, labels = equipamentos, autopct = '%1.1f%%', shadow = True, startangle = 90, 
        colors = sns.color_palette('muted'))

plt.axis('equal')
```
